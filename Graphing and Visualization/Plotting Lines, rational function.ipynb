```python
import matplotlib.pyplot as plt
```

```python
# y = mx +c

x = [-5,5]
m = 2
c =1

y =[340, 2345]

for i in range(0, len(x)):
    y[i] = m*x[i] +c
    
    
plt.plot(x,y, label ='y=%sx+%s')

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis =plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
```

```python
import numpy as np
```

```python
x =[-5,5]
m =[.7, -1.25]
b =[-2, .75]

for i in range(0, len(m)):
    y = m[i] * np.array(x) + b[i]
    plt.plot(x,y,label='%s' %i)
```

# Rational function

```python
x = range(-3,4)
y  = np.zeros(len(x))

print(x,y)
```

```python
for i in range(0, len(x)):
    y[i]=2- x[i]**2
    
plt.plot(x,y, 's-')
```

```python
x = np.linspace(-3,3,10)
y =2 + np.sqrt(abs(x))

plt.plot(x,y, 'ms-')
```

```python
x = np.linspace(-3,3,10)
y1 =2 + abs(x)**(1/3)

y2 =2 + abs(x)**(1/2)

plt.plot(x,y1, 'ms-')

plt.plot(x,y2, 'k-')
```
