```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
x =3
y =5
plt.plot(x,y, 'ro')
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()
plt.show()
```

```python
x = [-4,5,2,4,5,6,3]
y =[0,3,4,5,2,-3,4]

for i in range(0, len(x)):
    plt.plot(x[i],y[i], 'o', label ='point %s' %i)
    
    
plt.axis('square')
plt.grid()
plt.legend()
```

```python
import sympy as sym
```

```python
x = sym.symbols('x')
y = x**2 - 3*x
for pnt in range(-11,12):
    plt.plot(pnt, y.subs(x, pnt), 'o')
    
plt.show()
```

```python
p1= [-3,-1]
p2 = [4,4]

plt.plot(p1, p2) # well actaully this is not the line what we were expecting from points p1 and p2
```

```python
plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color=[.4, .2,.7], linewidth =5) # now that is the line we were looking for
plt.axis('square')
plt.axis([-6,6,-6,6])
```
