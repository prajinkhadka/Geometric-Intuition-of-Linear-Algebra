## Vector Addition : Geometrically

```python
# two vectors in R2

v1 = np.array([3,-1])
v2 = np.array([2,4])

v3 = v1+v2

#plotting

plt.plot([0, v1[0]], [0, v1[1]], 'b', label ='v1')
plt.plot([0, v2[0]] + v1[0], [0, v2[1]]+v1[1], 'r', label = 'v2')
plt.plot([0, v3[0]], [0, v3[1]], 'k', label ='v1+v2')

plt.legend()
plt.grid()
plt.show()
```

## Vector : Vector Multiplication ; dot product

```python
# there are many ways to perform the dot prodcut , lets look at some of thhem

v1 = np.array([1,2,3,4,5,6])
v2 = np.array([0,-4,-3,6,5, 1])

# method 1
dp = sum(np.multiply(v1, v2))

# method 2 
dp = np.dot(v1, v2)

#method 3
dp = np.matmul(v1, v2)

#method 4
dp =0 
for i in range(0, len(v1)):
    # multiply the coresponding elements and sum
    dp = dp + v1[i] * v2[i]
    
print(dp)
```

### length of vector

```python
# a vector
v1 = np.array([1,2,3,4,5,6])

# methods 1-4 just like with the regular dot product

vl = np.sqrt(sum(np.multiply(v1, v1)))

# meyhod 5: taking the norm

vl = np.linalg.norm(v1)

print(vl)
```

## dot product from Geometric perspective

```python
# two vectors

v1 = np.array([2,4,-3])
v2 = np.array([0, -3, -3])

# comppute the angle (radians ) bet ween vectors

ang = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))
print(ang)

# drawing them in 3d surface

fig = plt.figure()

ax = fig.gca(projection ='3d')

ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r')

print('angle between them is ', ang)
```

```python
# Equivvalence of algebraic and geometric dot product formulas, to show both are the same

# two vectors 

v1 = np.array([2,4, -3])
v2 = np.array([0, -3, -3])

# algebraic dot product

dp_a = np.dot(v1, v2)

# geometric is 

dp_g = np.linalg.norm(v1)* np.linalg.norm(v2) * np.cos(ang)

print(dp_a)
print(dp_g)
```

```python
# Outer Product of Vector

# two vectors
v1 = np.array([1,2,3])
v2 = np.array([-1, 0, 1])

# outer product

out = np.outer(v1, v2)
print(out)

# terrible programming
op = np.zeros((len(v1),len(v1)))
for i in range(0,len(v1)):
    for j in range(0,len(v2)):
        op[i,j] = v1[i] * v2[j]

print(op)
              
```

## Hermitian transpose : Conjugate Transpose

```python
# Complex Conjugate 
# complex Number : a + ib -> Complex Conjugate : a - ib
# complex Number : a - ib -> Complex Conjugate : a + ib
```

```python
# Now lets look at the Herimitian Transpose - > Its a normal transpose but with Conjugate transpose
# lets look at an example


# [1 + 3i, -2i, 4, -5] _> Herimitian transpose -> [1-3i
                                              #    +2i,
                                              #     4
                                              #     -5 ]
            
            
```

```python
# creating  a complex number
import numpy as np

z = np.complex(3,4)
print(z)

# magnitude 
print(np.linalg.norm(z))

# transpose 
print(np.transpose(z)*z)

# b herimitian transpose -> we we have the magnitude as 25, the normal transpose does not work in case of complex numbers
print(np.transpose(z.conjugate())*z)



# complex vectors

v = np.array([3,4j, 5+2j, np.complex(2,-5)])
print(v.T)
#transpose
print(np.transpose)

#conjugate transpose
print(np.transpose(v.conjugate()))
```
