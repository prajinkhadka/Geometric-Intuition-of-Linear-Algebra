```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.quiver(0,0,3,4)
plt.show()
```

```python
plt.grid()
plt.quiver(0,0,3,4, scale_units ='xy', angles = 'xy', scale =1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.grid()
plt.quiver(0,0,3,4, scale_units ='xy', angles = 'xy', scale =1, color='r')
plt.quiver(0,0,-3,4, scale_units ='xy', angles = 'xy', scale =1, color='b')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
def plot_vectors(vects):
    colors =['r', 'b', 'g', 'y']
    i =0
    for vec in vects:
        
        plt.quiver(vec[0],vec[1],vec[2],vec[3], scale_units ='xy', angles = 'xy', scale =1, color=colors[i%len(colors)])
        i += 1
    plt.xlim(-10, 10)
    plt.ylim(-10,10)

    
```

```python
plot_vectors([(0,0,3,4), (0,0,-3,4), (0,0,-3,2), (0,0,4,-1)])
```

```python
vecs = [(0,0,3,4), (0,0,-3,4), (0,0,-3,2), (0,0,4,-1)]
```

```python
plot_vectors([vecs[0], vecs[1]])
```
